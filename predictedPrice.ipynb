```python
#importing the libraries 
import pandas as pd
#reading the csv file
df = pd.read_csv('data/RiyadhAqqar.csv')
df.head()
```

```python
# حذف العمود الاول
df.drop("Unnamed: 0", axis=1, inplace=True)
 
df.head()
```

```python
from sklearn import linear_model

# استخراج البيانات من DataFrame
X = df[['Space', 'priceMeter']]
y = df['totalprice']

regr = linear_model.LinearRegression()
regr.fit(X, y)

# طلب إدخال القيم من المستخدم
Space = float(input("Enter Space: "))
priceMeter = float(input("Enter price of one meter: "))

# تنبؤ السعر الاجمالي
predictedPrice = regr.predict([[Space, priceMeter]])

print(predictedPrice)
```
